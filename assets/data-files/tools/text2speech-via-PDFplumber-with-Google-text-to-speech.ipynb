```python
import pdfplumber
from gtts import gTTS
import os
```

#Point to PDF file

```python
file_path = "Chapter 3.pdf"
```

#Extract pdf text

```python
text = ""

with pdfplumber.open(file_path) as pdf:
    for page in pdf.pages:
        text += page.extract_text() + "\n"
```

#Convert text to .mp3

```python
tts = gTTS(text, lang='en')
tts.save("Chapter 3.mp3")
```
